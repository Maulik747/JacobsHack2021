# Data Processing and Coordinate Transformation
When the device points some part of sky, the altitude and azimuth is calculated using gyroscope, accelerometer, and magnetometer sensors. 
Using astropy library, the coordinates of altitude and azimuth is transformed into RA Dec coordinates. Local-datetime, latitude, and longitude is required for this transformation. Local-datetime is obtained from the system and latitude and longitude is obtained either from GPS-module or inputed manually. 
With the obtained RA Dec coordinates, we query the database of astronomical catalogue to get the nearby astronomical objects (like stars, galaxies, nebulas, planets, etc.)

```python
import numpy as np
import astropy.units as u
from astropy.time import Time
from astropy.coordinates import SkyCoord, EarthLocation, AltAz
from datetime import datetime

from astroquery.simbad import Simbad
```

```python
#latitude and longitude of Jacobs University Bremen
#this data could be fed using GPS sensor from mobile device
lati =53.166158
long =8.650049
```

```python
#current time is the local time of the user
#function that transforms alt-azimuth coordinate to ICRS coordinate (RA, Dec)
def altAzmtoRaDec(azim, alti):
    jacobs = EarthLocation(lat=lati*u.deg, lon=long*u.deg, height=11*u.m)
    utcoffset = +1*u.hour #Central European Time, to be changed for another timezone
    current_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    time =Time(current_time) - utcoffset
    altaz = SkyCoord(az=azim*u.degree, alt=alti*u.degree, location=jacobs, obstime=time, frame='altaz')
    return(altaz.icrs)



```

```python
sample_coord = altAzmtoRaDec(245,37)

#function that returns the astronomical objects around a certain RA Dec coordinate
def listOfObjects(coords):
    print('function called')
    query_results = Simbad.query_region(coords, radius=0.5*u.degree)
    print(query_results)
    for i in query_results['MAIN_ID']:
        result = Simbad.query_objectids(i, verbose=False)
        if result is not None:
            obj_name = filter(lambda x: x != '*', result[0][0].split())
            prop_name = (' ').join(map(lambda x: x.capitalize(), obj_name))
                    # if a hd name capitalize hd
            if prop_name[:2].lower() == 'hd':
                prop_name = 'HD' + prop_name[2:]
            else:
                print("Not Found")
        print(prop_name)
        # for i in range(len(object_coords)):
        m_x = SkyCoord(ra='06d45m08.91728s', dec='-16d42m58.0171s', frame='icrs')
        m = m_x.info
        print(m)
# print(Simbad.query_objectids("IRAS 05396-0322"))
# print(result_table)
# for i in result_table['MAIN_ID']:
#     result_table = Simbad.query_objectids(i)
# print(result_table['MAIN_ID'])
# ls = Simbad()
# ls.ROW_LIMIT = 50 
# rt = ls.query_catalog('hip')
# x = np.array(rt['MAIN_ID'])
# print(np.char.strip(x, 'H'))
listOfObjects(sample_coord)
```
